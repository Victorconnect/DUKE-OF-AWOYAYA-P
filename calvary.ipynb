## WHAT IS A CONTEXT MANAGER?
A context manager is an object that defines a runtime context - it sets something up before a block of code runs, and tears it down when the block exists (even if it exists due to an exception). The with statement is the syntax uses them.

```python
file_path_txt = "Birthday_Wish_Template (1).txt"
file_path_csv = "products.csv"

with open(file_path_txt, "r", encoding="utf-8") as f:
    output = f.read()

print(output)
```

```python
def read_file(filepath: str) -> str:
    with open(filepath, "r", encoding="utf-8") as f:
        output = f.read()
    return output
```

```python
def modify(output: str, name: str, your_name: str) -> str:
    return output.replace("{name}", name).replace("[Your Name]", your_name)
```
